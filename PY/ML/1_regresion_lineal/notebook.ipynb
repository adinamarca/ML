## Regresión lineal

<img src="1.svg">



Recordemos que la regresión lineal, es un método mediante el cual buscamos ajustar la mejor recta posible a la nube de puntos constituida por los datos. En este sentido, trabajamos con datos cuantitativos.

El cómo realiza este ajuste es, comúnmente, a través del método de mínimos cuadrados ordinarios, ajustando el error de la recta a través de la modificación secuencial de parámetros, conocido como el descenso del gradiente, pero también, tenemos un método analítico.

## Algoritmos

Éstos métodos y su uso, varían en base a la cantidad de observaciones (datos) que tengamos. Por lo general, el descenso del gradiente es eficiente para muchas variables (n), mientras tanto, la ecuación normal es eficiente para un volumen bajo de variables (n < 10000).

### Descenso del gradiente

<img src="2.png">
<img src="3.png">

El proceso es a través del cálculo y la convergencia. 

Para los recién introducidos, el descenso del gradiente es, básicamente, un proceso iterativo. Partamos con los "errores": nosotros sabemos que, si nos equivocamos en algo, cada vez intentamos mejorar, ¿no? ¡Cada vez! Es decir, proceso tras proceso mejoramos: ITERANDO. Pero... ¿cómo lo hacemos en realidad? Definamos algunos conceptos:

• X: son nuestras variables de entrada. Con ellas, y a través de ellas, precisamos predecir.

• Θ: imagina tus primeros errores. Los theta, Θ, son los parámetros los cuales construirán la función que deseamos ajustar a la nube de puntos. Éstos valores sufrirán cambios a lo largo del proceso iterativo, hasta obtener la función ajustada.

• h(X): es nuestra función predictora.

• Y: necesitamos algo con qué comparar, algún resultado correcto, pues, estos son, justamente, nuestros valores Y. Son nuestros datos, con los cuales, intentaremos ajustar el mínimo error posible a ellos.

• J(Θ): es nuestra función de costos. Esta función tiene como objetivo calcular el error de nuestra recta. El error debe ser minimizado a través de la minimización de la sumatoria de las diferencias cuadradas entre Θ e Y, ¿por qué cuadradas? Es una forma de hacer que el error sea menos "tolerable", y por ende, que nuestra función de costos sea enormemente estricta con ellos. Al ser cuadrada, penalizaremos los errores con un alto valor, con mayor penalización, y aquellas con menor valor, con menor penalización. Dependiendo de la cantidad de parámetros que tengamos, la dimensionalidad de esta función aumentará. En el caso de la imagen de arriba, suponemos una sola variable, por lo que su minimización la encontraremos en el vértice de la parábola (mínimo global de la función), en donde coincide que, el parámetro Θ_1 deba ser 1, en orden para constituir, como hipótesis (al lado izquierdo) la función identidad, donde con la "nube" de tres puntos, obtenemos error cero al ser, la sumatoria cuadrada de las distancias verticales, entre los puntos y la función, 0.

• α: parámetro de aprendizaje. Se utiliza para controlar el avance de las iteraciones, los "pasos" a través del cuál cambian los parámetros Θ. Un valor muy alto puede significar una divergencia a los mínimos locales, mientras un valor bajo puede significar una ineficiencia en la iteración a la convergencia, realizando el proceso sumamente demoroso y costoso para la computadora.

Una vez, teniendo estos conceptos definidos, lo que se realiza es una diferencia entre los parámetros Θ actuales y el producto entre alpha y el gradiente del costo de forma de ir reduciendo, cada vez, la modificación de Θ, hasta converger las derivadas al mínimo cambio en un determinado número de iteraciones. Este determinado número de iteraciones se obtiene observando la gráfica de la función de costo, o por otros métodos. Una vez teniendo un mínimo cambio en las derivadas, la convergencia de nuestra función de costos está asegurada.

Nuevamente, con el gradiente de la función de costo se debe obtener el menor cambio en J de forma de obtener un <<theta>> convergente, por consecuencia de la convergencia al mínimo local o global (dependiendo de la función J).

Con el descenso de gradiente se obtiene una disminución automática del cambio en la derivada, siendo el cambio cada vez más pequeño a medida nos acercamos al mínimo local. Esto conlleva a que a la disminución del parámetro de aprendizaje no sea necesaria a lo largo del tiempo. Es decir, el gradiente de descenso da cada vez pasos más pequeños puesto la derivada converge a cero.

## Implementando la ecuación normal

Del gráfico observamos que, a medida aumenta el valor medio de la vivienda, también incrementa el número de habitaciones.

Ahora lo que realizaremos será <b>implementar la fórmula de la ecuación normal con código</b>.

Fórmula para minimizar el error cuadrático medio (MCO): $W = (X^{T}X)^{-1}X^{T}Y $

La ciencia es que las W las obtendremos manipulando nuestras matrices X e Y. Recordemos que no estamos trabajando de forma iterativa (para cada uno de los datos), sino de forma vectorial, con matrices que representan los datos de entrada y salida, operando de forma eficiente en nuestros computadores.

Hubo algo que no mencionamos, y es que existen dos tipos de parámetros:

• Pendientes $W_i$.

• Sesgo (bias) $W_0 $.

<img src="11.jpg">

Los sesgos sirven para cuando deseamos proporcionar cierto 'umbral' o manipulaciones a nuestras funciones. Un alto valor de bias puede perjudicar nuestro modelo, añadiendo un sesgo alto, por lo cual, nuestra predicción no será certera. En este sentido, es relevante tener un balance en este término.

<img src="12.png">

[Un enlace interesante sobre el bias, el underfitting, overfitting y cómo realizar un buen modelo de ML](https://towardsdatascience.com/bias-variance-and-regularization-in-linear-regression-lasso-ridge-and-elastic-net-8bf81991d0c5).

```python
# Matriz de entrada (número de habitaciones por hogares de cada barrio)
X = np.array(boston.data[:, 5])

# Añadimos columnas de 1s para término independiente (bias).
X = np.array([np.ones(X.shape[0]) ,X])
```

## Empezamos con Machine Learning sencillito
Escribiremos y programaremos $W = (X^{T}X)^{-1}X^{T}Y $.

```python
# Matriz de entrada (número de habitaciones por hogares de cada barrio)
X = np.array(boston.data[:, 5])

# Añadimos columnas de 1s para término independiente (bias).
X = np.array([np.ones(X.shape[0]) ,X])
X = X.T

# Con X.T definimos la traspuesta de X, que es más cómoda que np.transpose(X).
# W = (X.T * X): notar que con '*' hacemos la multiplicación normal, y nosotros necesitamos la multiplicación matricial con '@'.
# Con np.linalg.inv() invertimos la matriz.

W = np.linalg.inv(X.T @ X) @ X.T @ Y
W
```

## ¿Qué nos explica el modelo de regresión lineal simple?

Ahora tendríamos nuestra matriz de parámetros que minimizan el error cuadrático medio de nuestra nube de puntos. La matriz debería tener el elemento independiente y la pendiente.

• En este caso, el bias sería el coeficiente de posición ($-34.671$).

• La pendiente sería $9.1$. Ésto nos indica que, en un incremento de de una unidad del número de habitaciones, hay un incremento de 9.1 unidades el valor medio de la vivienda.


Ahora, si <b>graficamos<b/> nuestro modelo y la nube de puntos...

```python
X = np.array(boston.data[:, 5])
plt.scatter(X, Y, alpha=0.3)
# Definimos los dos puntos de la recta (inicial y final).
# Para ello, definimos que nuestro gráfico esté entre X e [4, 9]
# Los puntos del eje Y deberán ser dos (punto inicial y final de la recta de regresión obtenida por el modelo)
print(W[0], W[1])
plt.plot([4, 9], [W[0] + W[1]*4, W[0] + W[1]*9], c='red')
plt.show()
```

## Problema de la ecuación normal

La operación 'np,linalg.inv()' invierte una matriz. El inconveniente con ello es que es un procedimiento que, para el ordenador, requiere de n^3 cálculos, siendo n la dimensionalidad de la matriz, por lo que es sumamente importante ocupar la ecuación normal solo para ciertas situaciones, como por ejemplo, en las cuales <b>el número de características y observaciones no posee un gran volumen</b>, dado que de otra manera, será un proceso demoroso, en especial, para los <b>potato-core</b>.

Si suponemos 100.000 observaciones y 50 características (un problema más complejo), tendríamos $(100000 * 50)^{3}$ cálculos en el ordenador, lo que resulta en un sistema poco eficiente.

## Descenso del gradiente

Para las situaciones con enormes volúmenes de datos, requerimos de un método que no despeje exactamente el mínimo error de la función de coste, sino que de un método iterativo que se acerque poco a poco al mínimo error. Este algoritmo es el descenso del gradiente.
